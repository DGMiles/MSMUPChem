```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
n1=1
n2=2
L=1.0
x_array=np.linspace(0,L,100)
y_array=np.linspace(L,0,100)
x_grid,y_grid=np.meshgrid(x_array,y_array)
psi=(2/L)*np.sin(n1*np.pi*x_grid/L)*np.sin(n2*np.pi*y_grid/L)
fig = plt.figure()
ax = Axes3D(fig)
ax=fig.add_subplot(projection='3d')
ax.plot_surface(x_grid,y_grid,psi)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$\psi(x,y)$')
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax=fig.add_subplot(projection='3d')
ax.plot_surface(x_grid,y_grid,psi**2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$\psi^2(x,y)$')
```

```python

contours=plt.contourf(x_grid,y_grid,psi)
plt.xlabel('x')
plt.ylabel('y')
#clabel(contours)
fig.colorbar(contours)
plt.show()
```

```python

contours=plt.contourf(x_grid,y_grid,psi**2)
plt.xlabel('x')
plt.ylabel('y')
#clabel(contours)
fig.colorbar(contours)
plt.show()
```
