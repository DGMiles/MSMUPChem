```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import h #Planck's constant
from scipy.constants import c #speed of light
from scipy.constants import k #Boltzmann constant
from IPython.display import Markdown as md
from scipy.signal import find_peaks


def wavefcn(distance, n, L):
    psi = np.sin(n*np.pi*distance/L)
    return psi



#let's graph a particle in a box wavefunction
n=4.0       #an arbitrary quantum number
L=1.0       #an arbitrary length
increment=0.001
x=np.arange(0,L,increment)

psi1=wavefcn(x,n,L)
peaks = find_peaks(psi1)
plt.plot(x,psi1,'ro')
plt.grid()
plt.show()
print("peaks=",peaks[0]*increment)
valleys=find_peaks(-psi1)
print("valleys=",valleys[0]*increment)


psi2=psi1**2
plt.plot(x,psi2,'bo')
plt.grid()
plt.show()

peaks2 = find_peaks(psi2)
print("peaks2=",peaks2[0]*increment)
zeroes=find_peaks(-psi2)
print("zeroes=",zeroes[0]*increment)







```
