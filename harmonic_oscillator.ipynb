```python
# First load the numpy/scipy/matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
from scipy.signal import find_peaks
from math import factorial
from IPython.display import display




v=1       #an arbitrary vibrational quantum number
L=5.0       #an arbitrary length dimension
increment=0.001
x=np.arange(-L,L,increment)

def vib_psi(v,x):
    Hr=hermite(v)
    vib_psix = Hr(x)*np.exp(-0.5*x**2)
    return vib_psix


plt.plot(x,vib_psi(v,x),'ro')
plt.grid()
plt.show()
peaks = find_peaks(vib_psi(v,x))
print("peaks=",(peaks[0]-L/increment)*increment)
valleys=find_peaks(-vib_psi(v,x))
print("valleys=",(valleys[0]-L/increment)*increment)
```

```python
y=vib_psi(v,x)**2
plt.plot(x,y,'ro')
plt.grid()
plt.show()
peaks = find_peaks(y)
print("peaks=",(peaks[0]-L/increment)*increment)
valleys=find_peaks(-y)
print("valleys=",(valleys[0]-L/increment)*increment)
```
